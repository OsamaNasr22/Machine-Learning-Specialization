```python
import graphlab
```

```python
products = graphlab.SFrame('amazon_baby.gl/')
```

```python
products.head()
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love',
                  'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products ['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
products.head()
```

```python

def awesome_count(word_count):
    
    if selected_words[count] in word_count:
        return word_count[selected_words[count]]
    else:
        return 0
        
        
```

```python
count = 0
for word in selected_words:
    products[word] = products['word_count'].apply(awesome_count)
    count = count+1
```

```python
products.head()
```

```python
products['word_count'][0]
```

```python
for word in selected_words:
    print 'the word %s appeared %d times'%(word,products[word].sum())
```

```python
products = products[products['rating']!= 3]  
```

```python
products['sentiment'] = products['rating'] >= 4
```

```python
products.head()
```

```python
train_data, test_data = products.random_split(.8 , seed = 0)
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data
                                                          ,target = 'sentiment'
                                                          ,features = selected_words
                                                          ,validation_set = test_data)
```

```python
selected_words_model['coefficients'].print_rows(num_rows=12, num_columns=5)
```

```python
selected_words_model['coefficients'].sort('value', ascending = False).head()
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=['word_count'],
                                                     validation_set=test_data)
```

```python
selected_words_model.evaluate(test_data)
```

```python
sentiment_model.evaluate(test_data)
```

```python
products['predicted_sentiment'] = sentiment_model.predict(products, output_type='probability')
```

```python
products['predicted_selected_words'] = selected_words_model.predict(products, output_type='probability')
```

```python
products.head().sort('predicted_sentiment', ascending=False)
```
